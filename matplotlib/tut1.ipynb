1. <span style='font-family:"Times New Roman"'> <span style='color:Green'> **Creating simple plots**

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Creating a simple plot
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel('sin(x)')
plt.title('sine wave')
plt.grid()
plt.show()
```

```python
# Plot a cosine wave 
plt.plot(x,z,'r')
plt.xlabel("angle")
plt.ylabel('cosin(x)')
plt.title('cosine wave')
plt.grid()
plt.show()
```

```python
# Plotting a parabola 
x = np.linspace(-10,10,100)
y = x**2;

# Adding Markers an colors
plt.plot(x,y,'g.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.legend(['sin(x)'],loc = 'upper center')
plt.show()
```

2. <span style='font-family:"Times New Roman"'> <span style='color:Green'> **Creating Multiple plots**

```python
x = np.linspace(-10,10,10000)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.legend(['sin(x)','cos(x)'],loc = 'upper left')
plt.show()
```

3. <span style='font-family:"Times New Roman"'> <span style='color:Green'> **Creating Subplots**

```python
x = np.linspace(-10,10,1000)

# Plot 1 
plt.subplot(2,1,1)
plt.plot(x,np.sin(x),'g--')
plt.title('Sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# plot 2 
plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'r-.')
plt.title('Cosine wave')
plt.xlabel('x')
plt.ylabel('cos(x)')


plt.suptitle("Subplots")
plt.tight_layout()
plt.show()
```

4. <span style='font-family:"Times New Roman"'> <span style='color:Green'> **Bar plots**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
color = ['g','y','r','b','cyan']
ax.bar(languages,people,color=color)
plt.xlabel('LANGUAGE')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

5. <span style='font-family:"Times New Roman"'> <span style='color:Green'> **Pie Chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages,autopct='%1.1f%%')
plt.savefig("squares.tiff")
plt.show()
```

7. <span style='font-family:"Times New Roman"'> <span style='color:Green'> **Scatter Plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# 3D scatter plot
fig3 = plt.figure()
ax = fig3.add_subplot(111, projection='3d')
z = np.linspace(-10,10,100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x, y, z, c=z, cmap='Reds')
plt.show()
```
